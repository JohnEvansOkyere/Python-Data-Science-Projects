```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python

# Replace 'path_to_save/cleaned_data.csv' with the actual path to your CSV file
file_path =  "C:/Users/USER/QVI_data.csv"

# Read the CSV file into a new DataFrame
data = pd.read_csv(file_path)

# Now, 'new_data' contains the contents of your CSV file
# You can use 'new_data' in your new project or notebook
data.tail()
```

```python
# Set themes for plots
sns.set_theme(style="whitegrid")
plt.rcParams["axes.titlepad"] = 20  # Adjust the title padding

# Now you can use the 'data' DataFrame and create your plots using seaborn and matplotlib
```

```python
# Assuming you have a DataFrame named 'data' with a column 'DATE' containing the date information
data['DATE'] = pd.to_datetime(data['DATE'])  # Convert 'DATE' column to datetime format
data['Month_ID'] = data['DATE'].dt.strftime('%Y%m')  # Create a new column 'Month_ID'

# Now 'Month_ID' will have the format yyyymm
```

```python

# Specify the IDs you want to view as control stores
selected_ids = [77, 86, 88]

# Filter the DataFrame based on the specified control store IDs
controlStores = data[data['STORE_NBR'].isin(selected_ids)]

# Display the control stores
print("Control Stores:")
print(controlStores)

# Filter the DataFrame to get the remaining rows as trial stores
trialStores = data[~data['STORE_NBR'].isin(selected_ids)]

# Display the trial stores
print("\nTrial Stores:")
print(trialStores)
```
