```python
names = []
prices = []
rates = []
reviews = []
marks = []



url = "https://www.kilimall.co.ke/search-result?keyword=laptop&source=search|enterSearch|laptop"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

listings = soup.find_all("div", class_="info-box")

for i in listings:
    #Getting the names
    name = i.find("p", "class", class_="product-title").text
    names.append(name)

    #Getting the prices
    price = i.find("div", class_="product-price").text
    prices.append(price)

    rate = i.find("div", class_="rate").text
    rates.append(rate)

    review = i.find(class_="reviews").text
    reviews.append(review)

    mark = i.find("div", class_="mark-box").text
    marks.append(mark)

  

```

```python
data = pd.DataFrame({'name':names,
                     'price':prices,
                     'rate': rates,
                     'review': reviews,
                     "mark" : marks
                     })

data
```
