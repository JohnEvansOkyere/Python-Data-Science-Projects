```python
def first_approach():
    global df
    # And we'll just paste our code right here
    return (df.where(df['SUMLEV']==50)
             .dropna()
             .set_index(['STNAME','CTYNAME'])
             .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))

# Read in our dataset anew


# And now lets run it
timeit.timeit(first_approach, number=10)
```

```python
# Let's run this for once and see how fast it is
def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['TNAME','CTYNAME'], inplace=True)
    return new_df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})

# Read in our dataset anew

# And now lets run it
timeit.timeit(second_approach, number=10)
```

```python
def min_max(row):
    data = row[[
        'POPESTIMATE2010',
        'POPESTIMATE2011',
        'POPESTIMATE2012',
        'POPESTIMATE2013',
        'POPESTIMATE2014',         
        'POPESTIMATE2015'
    ]]
    
    return pd.Series({'min':np.min(data), 'max':np.max(data)})
```

```python
df = pd.read_csv('cwurData.csv')
df.head()
```

```python
def create_category(ranking):
    if (ranking >=1) & (ranking <=100):
        return "First Tier University "
    elif(ranking >=101) & (ranking <=200):
        return "Second Tier University"
    elif (ranking >=201)& (ranking <=200):
        return "Third Tier University"
    return "Other Top University"
    

df['Rank_Level'] = df['world_rank'].apply(lambda x:create_category(x))
df.head()
```

```python
df.pivot_table(values='score', index='country', columns = 'Rank_Level', aggfunc = [np.mean]).head()
```

```python
df.pivot_table(values='score', index = 'country', columns='Rank_Level',aggfunc=[np.mean, np.max]).head()
```

```python
df.pivot_table(values='score', index = 'country', columns='Rank_Level',aggfunc=[np.mean, np.max],margins=True).head()
```

```python
new_df = df.pivot_table(values='score', index = 'country', columns='Rank_Level',aggfunc=[np.mean, np.max]).head()

print(new_df.index)
print(new_df.columns)
```
