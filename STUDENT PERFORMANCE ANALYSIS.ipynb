#CORRELATION ANALYSIS

```python
#calculate the correlation matrix

cor_matrix = data.corr()

cor_matrix
```

```python
#visualizing the correlation matrix using heatmap

plt.figure(figsize=(10,8))
sns.heatmap(cor_matrix, annot = True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Correlation between 'study_hours' and 'grades'

studytime_grades_corr = data['studytime'].corr(data['grades'])
print(f"Correlation between study hours and grades: {studytime_grades_corr:.2f}")

"""In summary, a correlation coefficient of 0.25 suggests a positive but relatively weak correlation between study hours
and grades. It indicates that, on average, students who study more tend to have slightly higher grades, but the 
relationship is not extremely strong. Always interpret correlation in the context of your specific dataset and consider
additional factors that might contribute to the observed relationship."""
```

```python
data = data.sort_values(by='G1')

data
```

```python
# Assuming 'Grades' is the column representing student performance
performance_trends = data.groupby('grades')[['G1','G2','G3']].mean()

performance_trends.head(10)
```

```python


# Plotting
plt.figure(figsize=(10, 6))
plt.plot(performance_trends.index, performance_trends.values, marker='o')
plt.title('Student Performance Trends Over Time')
plt.xlabel('Academic Year or Semester')
plt.ylabel('Average Grades')
plt.legend(title=['G1','G2','G3'])
plt.grid(True)
plt.show()
```
