```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://www.airlinequality.com/"

data = requests.get(url)
data
```

```python
#CHECKING STATUS

data.status_code
```

```python
data.text
```

```python
soup = BeautifulSoup(data.text, "html.parser")
soup
```

```python
# Find the parent container with the class "col-content" (not "col-col-content")
reviews = soup.find_all("div", class_="col-content")  # Use find_all for multiple elements

# Initialize the list for storing data
data = []
for review in reviews:
    
    comment = review.find("div", class_="text_content")  # Use review.find here
    data.append(comment ) # Extract the text content and clean it

# Create a DataFrame
df = pd.DataFrame({"reviews": data})

df.tail()
```
