TRAIN AND TEST DATA

```python
#selecting the predicting target

#features column
X = titanic.drop('Survived', axis=1)

#targeted column
y = titanic['Survived']
```

```python
from sklearn.tree import DecisionTreeRegressor

logmodel= DecisionTreeRegressor(random_state=1)

logmodel.fit(X,y)
```

```python
print('predictions for thefirst 5 survivors')
print(X.head())
print('the predictions are')

print(logmodel.predict(X.head()))
```

```python
#mean absolute error

from sklearn.metrics import mean_absolute_error

predicted_survivals= logmodel.predict(X)
mean_absolute_error(y,predicted_survivals)
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.

X_train, X_test,y_train,y_test= train_test_split(X,y, test_size=0.3, random_state=1)

logmodel= DecisionTreeRegressor()
logmodel.fit(X_train,y_train)

val_predictions = logmodel.predict(X_train)
print(mean_absolute_error, y_train, val_predictions)
```
