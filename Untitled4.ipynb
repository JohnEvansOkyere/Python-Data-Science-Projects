```python
import requests
from bs4 import BeautifulSoup

url = "https://www.amazon.com/Giniros-Breathable-Lightweight-Sneakers-Sapphire/dp/B0BP2318TN/ref=sr_1_2_sspa?crid=260F4F4CPTO3K&dib=eyJ2IjoiMSJ9.oc6CNjXhd1MRv4JcUADHDoP0pP2-zJCL9lbtZXk4Ed5EUk2QTsvXKFhnsih0ywHo6euX7lRdyDQNngHi5gaTbT0G6RMubs2MSetiAAR1WH7lMK1ThBmu2J2twoMeY4u97lYmQ_TDsLPabNaNUdo8LFvNHjRac8r2dhmv3JjkN-lD1UzeiV-7LfRjaSJSq2DEgHJKOvWCh4x8kXSohijEdWqdDTPA1oIcwSqJ-YKuZ5B5J4cpmReni-jRn-QBAOL3emO9GxpGWGLVYJKMvyR0Hz85kE9u4KGI1EgE595Li_E.GNoBr4UwaXyww4P-_vSdzCOnWh9_Bi3qV0MXto5qNKA&dib_tag=se&keywords=sneakers+for+men&qid=1719487564&sprefix=sneakers%2Caps%2C565&sr=8-2-spons&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&psc=1"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
    "X-Amzn-Trace-Id": "Root=1-6679f5b3-258bfcf634a347d314af720c"
}

# Send a GET request to the URL
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, "html.parser")

# Scrape product details
try:
    title = soup.find(id="productTitle").get_text().strip()
except AttributeError:
    title = None

try:
    price = soup.find("span", {"class": "a-price-whole"}).get_text().strip()
    price_fraction = soup.find("span", {"class": "a-price-fraction"}).get_text().strip()
    price = f"{price}.{price_fraction}"
except AttributeError:
    price = None

try:
    rating = soup.find("span", {"class": "a-size-base a-color-base"}).get_text().strip()
except AttributeError:
    rating = None

try:
    review_count = soup.find("span", {"id": "acrCustomerReviewText"}).get_text().strip()
except AttributeError:
    review_count = None

try:
    availability = soup.find("div", {"id": "availability"}).find("span").get_text().strip()
except AttributeError:
    availability = None

# Print scraped data
print("Title:", title)
print("Price:", price)
print("Rating:", rating)
print("Review Count:", review_count)
print("Availability:", availability)
```

```python
import csv

header = ["Title", 'Price', 'Rating', 'Review Count','Availability']
data = [[title, price, rating,review_count, availability]]

with open('customerscraping.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)  # Write the header
    writer.writerows(data)   # Write the data
```
